```python
def partidos_count(data):
  list_partidos = []
  for orador in data['Orador']:
    try:
      partido_sep = orador.split(', ',1)[1]
    except:
      partido_sep = ''

    partido_sep = partido_sep.split('-',1)[0]

    list_partidos.append(partido_sep)
  data['Partido'] = list_partidos
  return data
```

```python
data_table = partidos_count(data_table)
data_table['Partido'].value_counts()
```

```python
def labeling(data,dict_pos_part):
  list_pos = []
  for i in data['Partido']:
    list_pos.append(dict_pos_part[i])
  data['Posicionamento politico'] = list_pos
  return data
```

```python
import os
```

```python
ids_list = []
discursos = []
labels = []

for txt in os.listdir("./dados"):
    try:
        with open("./dados" + "/" + txt, "r") as arquivo:
            texto = arquivo.read()

            discursos.append(texto)
            ids_list.append(txt.split('.')[0])
        arquivo.close()
    except:
        continue
    labels.append(data_table[data_table['Id_discurso'] == txt.split('.')[0]]['Posicionamento politico'].tolist()[0])
data = {'id': ids_list, 'discurso': discursos,'label':labels}
dataframe = pd.DataFrame.from_dict(data)
```
